## Poisson

In the late 1800s, he investigated accidental deaths by horse kick of soldiers in the Prussian army. He analyzed 20 years of data for 10 army corps, equivalent to 200 years of observations of one corps.

- Using <b>scipy.stats.expon()</b> and <b>numpy.random.exponential()</b>
    - What is the probability that he will be able to complete the trip without having to replace the car battery?
    - What is the probability of the battery failure during the trip?
    - Plot the probability distribution in both cases.
    - Calculate the mean and variance.
    - <b>What if the car already spent 2000 miles</b>. In this case, What will be the probability that he will be able to complete the trip without having to replace the car battery?

### Using SciPy

```python
from scipy.stats import expon
```

```python
lambda_=10000
cdf=expon.cdf(5000,loc=0,scale=lambda_)
x=np.arange(0,50000)
pdf=stats.expon.pdf(x,loc=0,scale=lambda_)
print(cdf)
prop=1-cdf
print(prop)
```

```python
expon.stats(scale=lambda_)
x=np.arange(0,50000)
```

```python
sns.lineplot(x=x,y=pdf)
```
